```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '../data/vehicles.csv'
df = pd.read_csv(file_path)
```

```python
print("First 5 rows of the dataset:")
display(df.head())
```

```python
print(f"\nDataset shape: {df.shape}")
```

```python
print("\nDataset Info:")
df.info(verbose=True, show_counts=True)
```

```python
columns_to_drop = [
    'url', 'region_url', 'image_url', 'description', 'county',
    'VIN', 'lat', 'long', 'posting_date'
]
```

```python
df_cleaned = df.drop(columns=columns_to_drop)
```

```python
print("Columns dropped. New shape:", df_cleaned.shape)
display(df_cleaned.head())
```

```python
df_cleaned = df_cleaned[df_cleaned['price'] > 0]
```

```python
df_cleaned.dropna(subset=['odometer'], inplace=True)
```

```python
df_cleaned.dropna(subset=['year'], inplace=True)
```

```python
print("Rows with 0 price or missing odometer/year removed.")
print("New shape:", df_cleaned.shape)
```

```python
categorical_cols = [
    'condition', 'cylinders', 'fuel', 'title_status',
    'transmission', 'drive', 'size', 'type', 'paint_color'
]
```

```python
for col in categorical_cols:
    mode_val = df_cleaned[col].mode()[0]
    df_cleaned[col].fillna(mode_val, inplace=True)
```

```python
print("Missing values after imputation:")
print(df_cleaned.isnull().sum())
```

```python
# Convert 'year' from float to integer
df_cleaned['year'] = df_cleaned['year'].astype(int)

# Convert 'odometer' from float to integer
df_cleaned['odometer'] = df_cleaned['odometer'].astype(int)

print("Data types after conversion:")
df_cleaned.info()
```

```python
current_year = 2025
df_cleaned['age'] = current_year - df_cleaned['year']

df_cleaned = df_cleaned[(df_cleaned['year'] >= 1990) & (df_cleaned['year'] <= current_year)]

print("Created 'age' column and filtered out unrealistic years.")
display(df_cleaned.head())
```

```python
from pandasql import sqldf
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

```python
# SQL query to get the top 15 manufacturers
query = """
SELECT
    manufacturer,
    COUNT(*) AS num_cars
FROM
    df_cleaned
GROUP BY
    manufacturer
ORDER BY
    num_cars DESC
LIMIT 15;
"""

top_manufacturers_sql = pysqldf(query)

display(top_manufacturers_sql)

plt.figure(figsize=(12, 8))
sns.barplot(data=top_manufacturers_sql, x='num_cars', y='manufacturer', palette='viridis')
plt.title('Top 15 Car Manufacturers in the Dataset (from SQL query)')
plt.xlabel('Number of Cars')
plt.ylabel('Manufacturer')
plt.show()
```

```python
# SQL query to get the average price for the top manufacturers
query = """
SELECT
    manufacturer,
    AVG(price) AS avg_price
FROM
    df_cleaned
WHERE
    manufacturer IN (SELECT manufacturer FROM top_manufacturers_sql)
GROUP BY
    manufacturer
ORDER BY
    avg_price DESC;
"""

avg_price_sql = pysqldf(query)

plt.figure(figsize=(12, 8))
sns.barplot(data=avg_price_sql, x='avg_price', y='manufacturer', palette='plasma')
plt.title('Average Price by Top 15 Manufacturers (from SQL query)')
plt.xlabel('Average Price ($)')
plt.ylabel('Manufacturer')
plt.show()
```

```python
# SQL query to get the average price by transmission type
query = """
SELECT
    transmission,
    AVG(price) AS avg_price,
    COUNT(*) AS num_listings
FROM
    df_cleaned
GROUP BY
    transmission;
"""

transmission_prices = pysqldf(query)

display(transmission_prices)

plt.figure(figsize=(8, 6))
sns.barplot(data=transmission_prices, x='transmission', y='avg_price')
plt.title('Average Price by Transmission Type (from SQL query)')
plt.xlabel('Transmission Type')
plt.ylabel('Average Price ($)')
plt.show()
```

```python
# SQL query to count cars by their condition
query = """
SELECT
    condition,
    COUNT(*) AS num_cars
FROM
    df_cleaned
GROUP BY
    condition
ORDER BY
    num_cars DESC;
"""

condition_counts = pysqldf(query)

# Visualize
plt.figure(figsize=(10, 6))
sns.barplot(data=condition_counts, x='num_cars', y='condition', palette='magma')
plt.title('Number of Cars by Condition (from SQL query)')
plt.xlabel('Number of Cars')
plt.ylabel('Condition')
plt.show()
```

```python
# SQL query to calculate average price and count for drive types
utility_query = """
SELECT
    drive,
    AVG(price) AS average_price,
    COUNT(*) AS number_of_listings
FROM
    df_cleaned
WHERE
    drive IN ('4wd', 'fwd')
GROUP BY
    drive;
"""

# Execute the query and display the results
utility_sql_results = pysqldf(utility_query)

print("Average Price and Counts by Drivetrain (from SQL query):")
display(utility_sql_results)
```

```python
from scipy import stats

# Prepare the data samples
four_wheel_drive_prices = df_cleaned[df_cleaned['drive'] == '4wd']['price']
front_wheel_drive_prices = df_cleaned[df_cleaned['drive'] == 'fwd']['price']

# Perform the two-sample t-test
t_stat_utility, p_value_utility = stats.ttest_ind(
    four_wheel_drive_prices,
    front_wheel_drive_prices,
    equal_var=False
)

# Set significance level and print conclusion
alpha = 0.05
print(f"\nT-statistic: {t_stat_utility:.4f}")
print(f"P-value: {p_value_utility}")

if p_value_utility < alpha:
    print("\nConclusion: Reject the Null Hypothesis.")
    print("There is a statistically significant difference in the average price between 4wd and fwd vehicles.")
else:
    print("\nConclusion: Fail to Reject the Null Hypothesis.")
    print("There is no statistically significant difference in price between 4wd and fwd vehicles.")
```

```python
# SQL query to calculate average price and count for transmission types
transmission_query = """
SELECT
    transmission,
    AVG(price) AS average_price,
    COUNT(*) AS number_of_listings
FROM
    df_cleaned
WHERE
    transmission IN ('automatic', 'manual')
GROUP BY
    transmission;
"""

# Execute the query and display the results
transmission_sql_results = pysqldf(transmission_query)

print("Average Price and Counts by Transmission (from SQL query):")
display(transmission_sql_results)
```

```python
# Prepare the data samples
automatic_prices = df_cleaned[df_cleaned['transmission'] == 'automatic']['price']
manual_prices = df_cleaned[df_cleaned['transmission'] == 'manual']['price']

# Perform the two-sample t-test
t_stat_trans, p_value_trans = stats.ttest_ind(
    automatic_prices,
    manual_prices,
    equal_var=False
)

# Set significance level and print conclusion
alpha = 0.05
print(f"\nT-statistic: {t_stat_trans:.4f}")
print(f"P-value: {p_value_trans}")

if p_value_trans < alpha:
    print("\nConclusion: Reject the Null Hypothesis.")
    print("There is a statistically significant difference in the average price between automatic and manual vehicles.")
else:
    print("\nConclusion: Fail to Reject the Null Hypothesis.")
    print("There is no statistically significant difference in price between automatic and manual vehicles.")
```
